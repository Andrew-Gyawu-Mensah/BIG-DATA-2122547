```python
import matplotlib.pyplot as plt # displaying images
import PIL # Python Imaging Library - image editing
import urllib #Library handling URLs (like requesting)
```

```python
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/4/4d/Cat_November_2010-1a.jpg/800px-Cat_November_2010-1a.jpg"
```

```python
image_days = urllib.request.urlopen(image_url)
image = PIL.Image.open(image_days)
np_image = np.array(image)
```

```python
plt.imshow(np_image)
```

```python
np_image.shape
```

```python
plt.imshow(np_image[100:500,150:500:,0:3])
```

```python
image_red_only = np_image.copy()
image_red_only[100:500,150:500,0] = 0 # Green Channel
image_red_only[100:500,150:500,2] = 0 # Blue Channel
plt.imshow(image_red_only)
```
